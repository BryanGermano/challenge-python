```
usuarios = []
grupos = []

desafios = desafios = [
   {
       "categoria": "Reciclagem",
       "descricao": "Separar resíduos recicláveis por 5 dias",
       "validado": False
   },
   {
       "categoria": "Limpeza de Praças",
       "descricao": "Participar da limpeza de uma praça local",
       "validado": False
   },
   {
       "categoria": "Economia de Energia",
       "descricao": "Desligar aparelhos sem uso durante uma semana",
       "validado": False
   },
   {
       "categoria": "Plantio de Árvores",
       "descricao": "Plantar uma árvore em local permitido",
       "validado": False
   },
   {
       "categoria": "Reciclagem",
       "descricao": "Separar plástico, papel e vidro corretamente",
       "validado": False
   },
   {
       "categoria": "Economia de Energia",
       "descricao": "Trocar lâmpadas comuns por LED\n",
       "validado": False

   }
]

categorias = (
   "Reciclagem",
   "Limpeza de Praças",
   "Economia de Energia",
   "Plantio de Árvores"
)

def validar_nome(nome):
   return len(nome.strip()) >= 3

def validar_idade(idade):
   if idade.isdigit():
       idade = int(idade)
       if 10 <= idade <= 100:
           return True
   return False

def validar_cpf(cpf):
   cpf = cpf.replace(".", "").replace("-", "")
   return cpf.isdigit() and len(cpf) == 11

def validar_opcao(op, minimo, maximo):
   if op.isdigit():
       op = int(op)
       if minimo <= op <= maximo:
           return True
   return False

def adicionar_pontos(usuario, pontos):
   usuario["pontos"] += pontos

def total_usuarios():
   return len(usuarios)

def cadastrar_usuario():
   print("\n CADASTRO DE USUÁRIO ")
   while True:
       nome = input("Nome: ")
       if validar_nome(nome):
           break
       print("Nome inválido")
   while True:
       idade = input("Idade: ")
       if validar_idade(idade):
           idade = int(idade)
           break
       print("Idade inválida")
   while True:
       cpf = input("CPF: ")
       if validar_cpf(cpf):
           break
       print("CPF inválido")
   usuario = {
       "nome": nome,
       "idade": idade,
       "cpf": cpf,
       "pontos": 0
   }
   usuarios.append(usuario)
   print("Usuário cadastrado com sucesso!")

def listar_usuarios():
   print("\n USUÁRIOS ")
   if len(usuarios) == 0:
       print("Nenhum usuário cadastrado")
       return
   for i, u in enumerate(usuarios, 1):
       print(
           f"{i}. {u['nome']} | "
           f"{u['idade']} anos | "
           f"CPF: {u['cpf']} | "
           f"{u['pontos']} pontos"
       )
   print(f"\nTotal: {total_usuarios()}")

def remover_usuario():
   print("\n REMOVER USUÁRIO ")
   if len(usuarios) == 0:
       print("Nenhum usuário cadastrado")
       return
   listar_usuarios()
   op = input("Escolha o usuário: ")
   if validar_opcao(op, 1, len(usuarios)):
       removido = usuarios.pop(int(op) - 1)
       print(f"{removido['nome']} removido!")
   else:
       print("Opção inválida")

def criar_grupo():
   print("\n CRIAR GRUPO ")
   if len(usuarios) == 0:
       print("Cadastre usuários antes")
       return
   nome_grupo = input("Nome do grupo: ")
   membros = []
   for i, u in enumerate(usuarios, 1):
       print(f"{i} - {u['nome']}")
   while True:
       op = input("Adicionar usuário (0 para sair): ")
       if op == "0":
           break
       if validar_opcao(op, 1, len(usuarios)):
           usuario = usuarios[int(op) - 1]
           if usuario not in membros:
               membros.append(usuario)
               print(f"{usuario['nome']} adicionado")
           else:
               print("Usuário já adicionado")
       else:
           print("Opção inválida")
   grupo = {
       "nome": nome_grupo,
       "membros": membros
   }
   grupos.append(grupo)
   print("Grupo criado com sucesso")

def listar_grupos():
   print("\n GRUPOS ")
   if len(grupos) == 0:
       print("Nenhum grupo criado")
       return
   for grupo in grupos:
       print(f"\nGrupo: {grupo['nome']}")
       if len(grupo["membros"]) == 0:
           print("Sem membros")
       else:
           for membro in grupo["membros"]:
               print("-", membro["nome"])

def criar_desafio():
   print("\n NOVO DESAFIO ")
   for i, c in enumerate(categorias, 1):
       print(f"{i} - {c}")
   while True:
       op = input("Categoria: ")
       if validar_opcao(op, 1, len(categorias)):
           categoria = categorias[int(op) - 1]
           break
       print("Categoria inválida")
   descricao = input("Descrição: ")
   desafio = {
       "categoria": categoria,
       "descricao": descricao,
       "validado": False
   }
   desafios.append(desafio)
   print("Desafio criado")

def listar_desafios():
   print("\n DESAFIOS ")
   if len(desafios) == 0:
       print("Nenhum desafio")
       return
   for i, d in enumerate(desafios, 1):
       status = "Validado" if d["validado"] else "Pendente"
       print(
           f"{i}. [{status}] "
           f"{d['categoria']} - "
           f"{d['descricao']}"
       )

def validar_desafio():
   print("\n VALIDAR DESAFIO ")
   if len(desafios) == 0:
       print("Sem desafios")
       return
   listar_desafios()
   op = input("Escolha o desafio: ")
   if validar_opcao(op, 1, len(desafios)):
       desafio = desafios[int(op) - 1]
       if desafio["validado"]:
           print("Desafio já validado")
           return
       desafio["validado"] = True
       for usuario in usuarios:
           adicionar_pontos(usuario, 10)
       print("Desafio validado")
       print("Todos ganharam 10 pontos")
   else:
       print("Opção inválida")


def ranking_usuarios():
   print("\n RANKING ")
   if len(usuarios) == 0:
       print("Nenhum usuário")
       return
   ranking = sorted(
       usuarios,
       key=lambda x: x["pontos"],
       reverse=True
   )
   for pos, u in enumerate(ranking, 1):
       print(f"{pos}º - {u['nome']} | {u['pontos']} pontos")

def menu():
   while True:

       print("1 - Cadastrar usuário")
       print("2 - Listar usuários")
       print("3 - Remover usuário")
       print("4 - Criar grupo")
       print("5 - Listar grupos")
       print("6 - Criar desafio")
       print("7 - Listar desafios")
       print("8 - Validar desafio")
       print("9 - Ranking")
       print("10 - Sair")
       op = input("Opção: ")
       if not validar_opcao(op, 1, 10):
           print("Opção inválida")
           continue
       op = int(op)
       match op:
           case 1:
               cadastrar_usuario()
           case 2:
               listar_usuarios()
           case 3:
               remover_usuario()
           case 4:
               criar_grupo()
           case 5:
               listar_grupos()
           case 6:
               criar_desafio()
           case 7:
               listar_desafios()
           case 8:
               validar_desafio()
           case 9:
               ranking_usuarios()
           case 10:
               print("Encerrando...")
               break

menu()
```
